```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("carprices.csv", names=['Car', 'Mileage', 'Price', 'Age'], skiprows=[0])

df
```

## Pandas get dummies

```python
dummies = pd.get_dummies(df.Car)

dummies
```

```python
dummies.drop('Mercedez Benz C class', axis='columns', inplace=True)

dummies
```

```python
merged_df = pd.concat([df, dummies], axis='columns')

merged_df
```

```python
merged_df.drop('Car', axis='columns', inplace=True)

merged_df
```

```python
features = merged_df.drop('Price', axis='columns').values

features
```

```python
target = merged_df.Price.values

target
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(features, target)
```

### 1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

```python
model.predict([[45000,4,0,0]])
```

### 2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[86000,7,0,1]])
```

### 3) Tell me the score (accuracy) of your model

```python
model.score(features, target)
```

## One Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
ohe_df = df

ohe_df
```

```python
ohe_df.Car = le.fit_transform(ohe_df.Car)

ohe_df
```

```python
X = ohe_df.drop('Price', axis='columns').values

X
```

```python
y = ohe_df.Price.values

y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([('Car', OneHotEncoder(), [0])], remainder='passthrough')
```

```python
X_ohe = ct.fit_transform(X)

X_ohe
```

```python
X_ohe = X_ohe[:,1:]

X_ohe
```

```python
model.fit(X_ohe, y)
```

```python
model.predict([[0,1,45000,4]])
```

```python
model.score(X_ohe, y)
```
