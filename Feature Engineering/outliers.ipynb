```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("AB_NYC_2019.csv")

df.head()
```

```python
df.price.describe()
```

```python
min_threshold = df.price.quantile(0.001)
max_threshold = df.price.quantile(0.9998)

min_threshold, max_threshold
```

```python
df[(df.price<min_threshold)]
```

```python
df[(df.price>max_threshold)]
```

```python
lower_bound, upper_bound = df.price.quantile([0.01, 0.999])

lower_bound, upper_bound
```

```python
final_df = df[(df.price>lower_bound) & (df.price<upper_bound)]

final_df.head()
```
